```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             mean_squared_error, mean_absolute_error, r2_score, silhouette_score, confusion_matrix)
from sklearn.cluster import KMeans, AffinityPropagation
import numpy as np
import matplotlib.pyplot as plt
from fpdf import FPDF
import seaborn as sns

# Caminhos dos datasets
caminho_agrupamento = 'datasets/agrupamento/X_dataset.csv'
caminho_classificacao = {
    'X_train': 'datasets/classificacao/X_training.csv',
    'y_train': 'datasets/classificacao/y_training.csv',
    'X_validation': 'datasets/classificacao/X_validation.csv',
    'y_validation': 'datasets/classificacao/y_validation.csv',
    'X_test': 'datasets/classificacao/X_test.csv',
    'y_test': 'datasets/classificacao/y_test.csv'
}
caminho_regressao = {
    'X_train': 'datasets/regressao/X_training.csv',
    'y_train': 'datasets/regressao/y_training.csv',
    'X_validation': 'datasets/regressao/X_validation.csv',
    'y_validation': 'datasets/regressao/y_val.csv',
    'X_test': 'datasets/regressao/X_test.csv',
    'y_test': 'datasets/regressao/y_test.csv'
}
```

```python
# Leitura dos arquivos de agrupamento
dados_agrupamento = pd.read_csv(caminho_agrupamento)

# Leitura dos arquivos de classificação
dados_classificacao = {key: pd.read_csv(path) for key, path in caminho_classificacao.items()}

# Leitura dos arquivos de regressão
dados_regressao = {key: pd.read_csv(path) for key, path in caminho_regressao.items()}
```

```python
# Funções para calcular as métricas
def calcular_metricas_classificacao(X_train, y_train, X_val, y_val, X_test, y_test):
    modelos = {
        'KNN': KNeighborsClassifier(),
        'Decision Tree': DecisionTreeClassifier(),
        'Random Forest': RandomForestClassifier(),
        'Logistic Regression': LogisticRegression(max_iter=1000)
    }
    
    resultados_treino = []
    resultados_validacao = []
    resultados_teste = []

    for nome, modelo in modelos.items():
        modelo.fit(X_train, y_train)

        # Dados de treinamento
        y_pred_train = modelo.predict(X_train)
        resultados_treino.append({
            'Algoritmo': nome,
            'Accuracy': accuracy_score(y_train, y_pred_train),
            'Precision': precision_score(y_train, y_pred_train, average='weighted', zero_division=0),
            'Recall': recall_score(y_train, y_pred_train, average='weighted'),
            'F1-Score': f1_score(y_train, y_pred_train, average='weighted'),
        })

        # Dados de validação
        y_pred_val = modelo.predict(X_val)
        resultados_validacao.append({
            'Algoritmo': nome,
            'Accuracy': accuracy_score(y_val, y_pred_val),
            'Precision': precision_score(y_val, y_pred_val, average='weighted', zero_division=0),
            'Recall': recall_score(y_val, y_pred_val, average='weighted'),
            'F1-Score': f1_score(y_val, y_pred_val, average='weighted'),
        })

        # Dados de teste
        y_pred_test = modelo.predict(X_test)
        resultados_teste.append({
            'Algoritmo': nome,
            'Accuracy': accuracy_score(y_test, y_pred_test),
            'Precision': precision_score(y_test, y_pred_test, average='weighted', zero_division=0),
            'Recall': recall_score(y_test, y_pred_test, average='weighted'),
            'F1-Score': f1_score(y_test, y_pred_test, average='weighted'),
        })

    return pd.DataFrame(resultados_treino), pd.DataFrame(resultados_validacao), pd.DataFrame(resultados_teste)
```

```python
def calcular_metricas_regressao(X_train, y_train, X_val, y_val, X_test, y_test):
    modelos = {
        'Linear Regression': LinearRegression(),
        'Decision Tree Regressor': DecisionTreeRegressor(),
        'Random Forest Regressor': RandomForestRegressor(),
        'Linear Regression Lasso': Lasso(),
        'Linear Regression Ridge': Ridge(),
        'Linear Regression Elastic Net': ElasticNet()
    }

    resultados_treino = []
    resultados_validacao = []
    resultados_teste = []

    for nome, modelo in modelos.items():
        modelo.fit(X_train, y_train)

        for conjunto, X, y in zip(['Treinamento', 'Validação', 'Teste'], 
                                   [X_train, X_val, X_test], 
                                   [y_train, y_val, y_test]):
            y_pred = modelo.predict(X)

            r2 = r2_score(y, y_pred)
            mse = mean_squared_error(y, y_pred)
            rmse = np.sqrt(mse)
            mae = mean_absolute_error(y, y_pred)
            mape = np.mean(np.abs((y - y_pred) / y)) * 100

            if conjunto == 'Treinamento':
                resultados_treino.append({
                    'Algoritmo': nome,
                    'R2': r2,
                    'MSE': mse,
                    'RMSE': rmse,
                    'MAE': mae,
                    'MAPE': mape
                })
            elif conjunto == 'Validação':
                resultados_validacao.append({
                    'Algoritmo': nome,
                    'R2': r2,
                    'MSE': mse,
                    'RMSE': rmse,
                    'MAE': mae,
                    'MAPE': mape
                })
            elif conjunto == 'Teste':
                resultados_teste.append({
                    'Algoritmo': nome,
                    'R2': r2,
                    'MSE': mse,
                    'RMSE': rmse,
                    'MAE': mae,
                    'MAPE': mape
                })

    return pd.DataFrame(resultados_treino), pd.DataFrame(resultados_validacao), pd.DataFrame(resultados_teste)

def calcular_metricas_agrupamento(X):
    modelos = {
        'K-Means': KMeans(n_clusters=8),
        'Affinity Propagation': AffinityPropagation()
    }

    resultados = []
    for nome, modelo in modelos.items():
        modelo.fit(X)
        labels = modelo.labels_

        silhouette_avg = silhouette_score(X, labels)
        
        if nome == 'K-Means':
            num_clusters = modelo.n_clusters
        elif nome == 'Affinity Propagation':
            num_clusters = len(set(labels))

        resultados.append({
            'Algoritmo': nome,
            'Silhouette Score': silhouette_avg,
            'Número de Clusters': num_clusters
        })

        plt.figure(figsize=(8, 6))
        plt.title(f"Agrupamento com {nome}")
        plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis')
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.colorbar(label='Cluster')
        plt.savefig(f"{nome.replace(' ', '_')}_agrupamento.png")
        plt.close()

    return pd.DataFrame(resultados)
```

```python
# Função para gerar um PDF com os resultados
def gerar_pdf(resultados_classificacao, resultados_regressao, resultados_agrupamento):
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()
    

    # Título
    pdf.set_font("Arial", 'B', 16)
    pdf.cell(0, 5, 'Resultados de Machine Learning', ln=True, align='C')
    pdf.cell(0, 5, ln=True)

    # Resultados de Classificação
    pdf.set_font("Arial", 'B', 14)
    pdf.cell(0, 5, ln=True)
    pdf.cell(0, 5, 'Resultados de Classificação', ln=True)

    pdf.set_font("Arial", '', 12)
    for index, row in resultados_classificacao.iterrows():
        pdf.cell(0, 5, f"{row['Algoritmo']}: Accuracy = {row['Accuracy']:.2f}, Precision = {row['Precision']:.2f}, Recall = {row['Recall']:.2f}, F1-Score = {row['F1-Score']:.2f}", ln=True)

    #pdf.add_page()  # Nova página para os resultados de regressão
    print("\n")

    # Resultados de Regressão
    pdf.set_font("Arial", 'B', 14)
    pdf.cell(0, 5, ln=True)
    pdf.cell(0, 5, 'Resultados de Regressão', ln=True)
    

    pdf.set_font("Arial", '', 12)
    for index, row in resultados_regressao.iterrows():
        pdf.cell(0, 5, f"{row['Algoritmo']}: R2 = {row['R2']:.2f}, MSE = {row['MSE']:.2f}, RMSE = {row['RMSE']:.2f}, MAE = {row['MAE']:.2f}, MAPE = {row['MAPE']:.2f}", ln=True)

    #pdf.add_page()  # Nova página para os resultados de agrupamento
    print("\n")

    # Resultados de Agrupamento
    pdf.cell(0, 5, ln=True)
    pdf.set_font("Arial", 'B', 14)
    pdf.cell(0, 5, 'Resultados de Agrupamento', ln=True)

    pdf.set_font("Arial", '', 12)
    for index, row in resultados_agrupamento.iterrows():
        pdf.cell(0, 5, f"{row['Algoritmo']}: Silhouette Score = {row['Silhouette Score']:.2f}, Número de Clusters = {row['Número de Clusters']}", ln=True)

    # Adiciona imagens dos agrupamentos
    #pdf.add_page()
    pdf.cell(0, 5, ln=True)
    pdf.set_font("Arial", 'B', 14)
    pdf.cell(0, 5, 'Visualizações de Agrupamento', ln=True)

    # Adicionar as imagens usando os nomes corretos
    for nome in ['K-Means', 'Affinity Propagation']:
        try:
            pdf.image(f"{nome.replace(' ', '_')}_agrupamento.png", w=180)
            pdf.cell(0, 5, '', ln=True)  # Espaço entre as imagens
        except RuntimeError as e:
            print(f"Erro ao adicionar a imagem {nome}: {e}")

    # Salvar PDF
    pdf.output("resultados_maquina.pdf")
```
